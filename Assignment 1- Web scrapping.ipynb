```python
from bs4 import BeautifulSoup
import requests

page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page

soup = BeautifulSoup(page.content)
soup

first_title = soup.find('div',class_='gsc_mvt_table_wrapper')
first_title

first_title.text.split()

loc = soup.find('div', class_="gsc_mvt_table_wrapper")
loc.text

sta = soup.find('div',class_="gsc_mvt_table_wrapper")
sta

titles = []
for i in soup.find_all('div', class_="gsc_mp_anchor_lrge"):
    titles.append(i.text.split())

titles

location = []
for i in soup.find_all('div', class_="gsc_mvt_table_wrapper"):
    location.append(i.text)
    
location
```
